```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from skimage import io
```

```python
# Load image
image = io.imread('photo-1497671954146-59a89ff626ff.jfif')
image = image / 255.0  # Normalize pixel values (optional)

# Reshape image to a 2D array of pixels (n_samples, n_features)
pixels = image.reshape(-1, 3)
```

```python
# Apply KMeans
k = 16  # Number of colors (clusters)
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(pixels)
```

```python
# Replace each pixel with its nearest centroid
compressed_pixels = kmeans.cluster_centers_[kmeans.labels_]
compressed_image = compressed_pixels.reshape(image.shape)
```

```python
# Show original vs compressed
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title(f"Compressed Image with {k} colors")
plt.imshow(compressed_image)
plt.axis('off')

plt.show()
```
